```python
import os
import random
import matplotlib.pyplot as plt
from IPython.display import Image

def childrens_generative_ABC_book():
    # Specify the directory containing the images
    image_directory = "c:\\working_directory\\ABC_images\\"

    # Prompt the user for a letter (case-insensitive)
    letter = input("Enter a letter: ").lower()

    # Get a list of image filenames that start with the given letter
    matching_files = [filename for filename in os.listdir(image_directory) if filename.lower().startswith(letter)]

    if not matching_files:
        print(f"No images found starting with letter '{letter}'.")
    else:
        # Select a random image filename
        selected_filename = random.choice(matching_files)
        image_path = os.path.join(image_directory, selected_filename)

        # Extract the image name (without extension) for the caption
        image_name = os.path.splitext(selected_filename)[0]

        # Display the image
        img = Image(filename=image_path)
        display(img)

        # Generate and display the caption
        caption = f"{letter.upper()}{letter} is for {image_name}"
        print(caption)

# Example usage:
# Run the function to prompt for a letter and display a random image with a caption
childrens_generative_ABC_book()
```

```python
from PIL import Image, ImageDraw

def generate_cartoon_image(subject, image_directory):
    # Create a white background image (2x3 size)
    width, height = 2 * 100, 3 * 100  # Adjust the size as needed
    background_color = (255, 255, 255)  # White color
    img = Image.new("RGB", (width, height), background_color)
    draw = ImageDraw.Draw(img)

    # Draw your subject (e.g., cat or cow)
    # You can customize this based on your desired cartoon style
    if subject.lower() == "cat":
        # Draw a simple cat-like shape (example)
        draw.ellipse((50, 50, 150, 150), fill=(255, 0, 0))  # Red cat head
        draw.ellipse((75, 75, 100, 100), fill=(0, 0, 0))  # Black cat eyes
        draw.ellipse((125, 75, 150, 100), fill=(0, 0, 0))  # Black cat eyes
        draw.arc((80, 90, 140, 140), start=0, end=180, fill=(0, 0, 0), width=3)  # Cat smile
    elif subject.lower() == "cow":
        # Draw a simple cow-like shape (example)
        draw.ellipse((50, 50, 150, 150), fill=(255, 255, 255))  # White cow head
        draw.rectangle((75, 100, 125, 150), fill=(0, 0, 0))  # Black cow spots
        draw.ellipse((80, 75, 100, 95), fill=(0, 0, 0))  # Black cow eyes
        draw.ellipse((120, 75, 140, 95), fill=(0, 0, 0))  # Black cow eyes

    # Save the image to the specified directory
    image_path = os.path.join(image_directory, f"{subject.lower()}_cartoon.png")
    img.save(image_path)

    # Display the generated image
    img.show()

# Example usage:
# generate_cartoon_image("cat", "c:\\working_directory\\ABC_images\\")
# generate_cartoon_image("cow", "c:\\working_directory\\ABC_images\\")
```
